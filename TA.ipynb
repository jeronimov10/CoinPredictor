## SMA

### Simple moving avarage - Es el promedio móvil simple de los últimos 50 días


sirve para:

1) Identificar tendencia: **Si el precio está por encima del SMA50, la tendencia es alcista. Si está por debajo, es bajista.**

2) Soporte/Resistencia: El SMA50 actúa como un nivel donde el precio tiende a rebotar.
Señales de entrada/salida:

    2.1. Cuando el precio cruza hacia arriba el SMA50 = señal de compra

    2.2. Cuando cruza hacia abajo = señal de venta


3) Cruces con otros SMA: Por ejemplo, cuando el SMA20 (más rápido) cruza por encima del SMA50 (más lento) es una señal alcista conocida como "Golden Cross".

4) SMA - x
    * SMA20 = corto plazo
    * SMA50 = mediano plazo (el que mencionas)
    * SMA200 = largo plazo

```python
df_prueba.ta.sma(lenght = 50)
```

```python
ta.sma(df_prueba['Close'], length = 50)
```

## RSI

### Relative Strength Index (Índice de Fuerza Relativa).

Es un oscilador que mide la velocidad y magnitud de los cambios de precio para identificar si un activo está sobrecomprado o sobrevendido.

* RSI va de 0 a 100

Sirve para:

1. Identificar sobrecompra/sobreventa:

    RSI > 70: Sobrecomprado → posible corrección a la baja (señal de venta)
    RSI < 30: Sobrevendido → posible rebote al alza (señal de compra)

2. Divergencias:

    Si el precio hace máximos más altos pero el RSI hace máximos más bajos = debilidad, posible caída
    Si el precio hace mínimos más bajos pero el RSI hace mínimos más altos = fortaleza, posible subida

3. Confirmación de tendencia:

    En tendencia alcista, RSI suele mantenerse entre 40-90
    En tendencia bajista, RSI suele mantenerse entre 10-60

```python
df_prueba.ta.rsi(lenght = 50)
```

## MACD

### Moving Average Convergence Divergence (Convergencia/Divergencia de Medias Móviles).
Es un indicador de momentum que muestra la relación entre dos medias móviles exponenciales (EMA) para identificar cambios de tendencia.

1. Cruces de líneas (señales de compra/venta):

    MACD cruza hacia arriba la línea de señal = señal de COMPRA (momentum alcista)
    MACD cruza hacia abajo la línea de señal = señal de VENTA (momentum bajista)

2. Cruce del cero:

    MACD cruza hacia arriba el cero = tendencia alcista confirmada
    MACD cruza hacia abajo el cero = tendencia bajista confirmada

3. Divergencias:

    Precio hace máximos más altos pero MACD hace máximos más bajos = debilidad, posible caída
    Precio hace mínimos más bajos pero MACD hace mínimos más altos = fortaleza, posible subida

4. Histograma:

    Barras creciendo = momentum aumentando
    Barras decreciendo = momentum debilitándose

```python
df_prueba.ta.macd(fast = 8, slow = 21)
```

### Resumen
1) SMA50 (Simple Moving Average 50):

    * Precio > SMA50 → Tendencia alcista, considera comprar o mantener
    * Precio < SMA50 → Tendencia bajista, considera vender o no entrar
    * Precio toca el SMA50 → Actúa como soporte (en subida) o resistencia (en bajada)

2) RSI (Relative Strength Index):

    * RSI > 70 → Sobrecomprado, el activo subió mucho, probable corrección a la baja (vender)
    * RSI < 30 → Sobrevendido, el activo cayó mucho, probable rebote al alza (comprar)
    * RSI entre 30-70 → Zona neutral, sin señales claras

3) MACD:

    * MACD > 0 → Momentum alcista
    * MACD < 0 → Momentum bajista
    * MACD cruza hacia arriba la línea de señal → Señal de COMPRA
    * MACD cruza hacia abajo la línea de señal → Señal de VENTA
    * Histograma creciendo → Fuerza aumentando
    * Histograma decreciendo → Fuerza debilitándose

## Estrategia

### ESTRATEGIA PARA SHORTS (Detectar Caídas)

### Objetivo
Identificar momentos donde el precio está a punto de BAJAR para abrir posiciones cortas (shorts) y ganar con la caída.

---

### Indicadores Usados

#### 1. **SMA 50** (Media Móvil Simple 50 días)
- **Si Precio < SMA50** ➜ Tendencia bajista confirmada
- Actúa como resistencia: cuando el precio sube y toca el SMA50, tiende a rebotar hacia abajo

#### 2. **EMA 20** (Media Móvil Exponencial 20 días)
- Más sensible que el SMA, detecta cambios rápidos
- **Si EMA20 < SMA50** ➜ "Death Cross" = señal bajista muy fuerte

#### 3. **RSI 14** (Relative Strength Index)
- **Si RSI > 70** ➜ Sobrecomprado, mucha gente ya compró, probable corrección a la baja
- Ideal para entrar en short cuando está en zona alta

#### 4. **MACD** (Moving Average Convergence Divergence)
- **Si MACD < Señal** ➜ Momentum bajista, la fuerza de compra se está debilitando
- El cruce hacia abajo es señal de venta

#### 5. **Bandas de Bollinger** (20 días, 2 desviaciones)
- **Si Precio > Banda Superior** ➜ Precio estirado hacia arriba, probable retroceso
- Cuando toca la banda superior en sobrecompra, tiende a volver al centro

#### 6. **ATR 14** (Average True Range)
- Mide volatilidad
- Te ayuda a calcular dónde poner el stop loss (ATR × 2 por encima del precio de entrada)

#### 7. **OBV** (On Balance Volume)
- Confirma con volumen
- Si precio sube pero OBV baja = divergencia = debilidad = probable caída

---

### Señal SHORT (Puntuación 0-5)

La columna Señal_SHORT suma cuántas condiciones bajistas se cumplen:

- **5 puntos** = SEÑAL MUY FUERTE de caída inminente 
- **4 puntos** = Señal fuerte, considerar short
- **3 puntos** = Señal moderada, esperar confirmación
- **0-2 puntos** = Sin señal clara o tendencia alcista

---

### Validación de la Estrategia

- **Cambio_5dias**: % de cambio del precio 5 días después
- **Cambio_10dias**: % de cambio del precio 10 días después

Si Cambio_5dias es **negativo** cuando Señal_SHORT = 5, la estrategia funcionó

---

### Ejemplo de Lectura

| Señal_SHORT | RSI_14 | MACD vs Señal | Precio vs SMA50 | Cambio_5dias | ¿Qué hacer? |
|-------------|--------|---------------|-----------------|--------------|-------------|
| 5           | 78     | MACD < Señal  | Precio < SMA50  | -3.2%        | SHORT perfecto |
| 3           | 65     | MACD < Señal  | Precio > SMA50  | +1.5%        | Señal mixta |
| 1           | 45     | MACD > Señal  | Precio > SMA50  | +5.0%        | No hacer short |

---

```python
short_strategy = ta.Strategy(
    name = 'Short_Signals',
    ta = [
        {'kind': 'sma', 'length': 50},      # Tendencia mediano plazo
        {'kind': 'ema', 'length': 20},      # Tendencia corto plazo
        {'kind': 'rsi', 'length': 14},      # Detectar sobrecompra
        {'kind': 'macd', 'fast': 12, 'slow': 26, 'signal': 9},  # Momentum
        {'kind': 'bbands', 'length': 20},   # Volatilidad y límites
        {'kind': 'atr', 'length': 14},      # Volatilidad para stop loss
        {'kind': 'obv'},                    # Volumen
    ]
)


ticker = 'BTC-USD'
data = yf.Ticker(ticker).history(period='1y')


data.ta.strategy(short_strategy)

# Calcular señales
data['Precio_debajo_SMA50'] = (data['Close'] < data['SMA_50']).astype(int)
data['EMA20_debajo_SMA50'] = (data['EMA_20'] < data['SMA_50']).astype(int)
data['RSI_sobrecompra'] = (data['RSI_14'] > 70).astype(int)
data['MACD_bajista'] = (data['MACD_12_26_9'] < data['MACDs_12_26_9']).astype(int)
data['Precio_cerca_BB_superior'] = (data['Close'] > data['BBU_20_2.0']).astype(int)


data['Señal_SHORT'] = (
    data['Precio_debajo_SMA50'] + 
    data['EMA20_debajo_SMA50'] + 
    data['RSI_sobrecompra'] + 
    data['MACD_bajista'] + 
    data['Precio_cerca_BB_superior']
)


data['Cambio_5dias'] = ((data['Close'].shift(-5) - data['Close']) / data['Close'] * 100).round(2)
data['Cambio_10dias'] = ((data['Close'].shift(-10) - data['Close']) / data['Close'] * 100).round(2)


columnas_importantes = ['Close', 'SMA_50', 'EMA_20', 'RSI_14', 'MACD_12_26_9', 
                        'MACDs_12_26_9', 'Señal_SHORT', 'Cambio_5dias', 'Cambio_10dias']
```

```python
import matplotlib.pyplot as plt


señales_fuertes = data[data['Señal_SHORT'] >= 3].copy()

print("=" * 70)
print(f"ANÁLISIS DE SEÑALES SHORT PARA {ticker}")
print("=" * 70)
print(f"\nSeñales fuertes encontradas (≥3 puntos): {len(señales_fuertes)}")

if len(señales_fuertes) > 0:
    print("\nÚLTIMAS 10 SEÑALES FUERTES:")
    print(señales_fuertes[['Close', 'RSI_14', 'MACD_12_26_9', 'Señal_SHORT', 'Cambio_5dias', 'Cambio_10dias']].tail(10).to_string())
    

    efectivas = (señales_fuertes['Cambio_5dias'] < 0).sum()
    total = len(señales_fuertes[señales_fuertes['Cambio_5dias'].notna()])
    tasa_exito = (efectivas / total * 100) if total > 0 else 0
    
    print(f"\nEfectividad de la estrategia: {tasa_exito:.1f}%")
    print(f"   ({efectivas} de {total} señales resultaron en caídas)")
    

    caidas = señales_fuertes[señales_fuertes['Cambio_5dias'] < 0]['Cambio_5dias']
    if len(caidas) > 0:
        print(f"\nCaída promedio cuando funciona: {caidas.mean():.2f}%")
        print(f"   Caída máxima capturada: {caidas.min():.2f}%")

# Gráfico: Precio + SMA50 + Señales SHORT
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# Subplot 1: Precio y SMA50
ax1.plot(data.index, data['Close'], label='Precio', linewidth=1.5)
ax1.plot(data.index, data['SMA_50'], label='SMA 50', linestyle='--', alpha=0.7)
ax1.plot(data.index, data['EMA_20'], label='EMA 20', linestyle='--', alpha=0.7)
ax1.scatter(señales_fuertes.index, señales_fuertes['Close'], color='red', s=100, marker='v', label='Señal SHORT', zorder=5)
ax1.set_ylabel('Precio (USD)')
ax1.set_title(f'{ticker} - Estrategia SHORT')
ax1.legend()
ax1.grid(alpha=0.3)

# Subplot 2: RSI
ax2.plot(data.index, data['RSI_14'], label='RSI 14', color='purple')
ax2.axhline(70, color='red', linestyle='--', alpha=0.5, label='Sobrecompra (70)')
ax2.axhline(30, color='green', linestyle='--', alpha=0.5, label='Sobreventa (30)')
ax2.fill_between(data.index, 70, 100, alpha=0.1, color='red')
ax2.set_ylabel('RSI')
ax2.legend()
ax2.grid(alpha=0.3)

# Subplot 3: MACD
ax3.plot(data.index, data['MACD_12_26_9'], label='MACD', color='blue')
ax3.plot(data.index, data['MACDs_12_26_9'], label='Señal', color='orange')
ax3.bar(data.index, data['MACDh_12_26_9'], label='Histograma', alpha=0.3)
ax3.axhline(0, color='black', linestyle='-', alpha=0.3)
ax3.set_ylabel('MACD')
ax3.set_xlabel('Fecha')
ax3.legend()
ax3.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Señal actual
print("\n" + "=" * 70)
print("SEÑAL ACTUAL (ÚLTIMOS 5 DÍAS):")
print("=" * 70)
print(data[['Close', 'RSI_14', 'MACD_12_26_9', 'Señal_SHORT']].tail(5).to_string())

ultima_señal = data['Señal_SHORT'].iloc[-1]
if ultima_señal >= 4:
    print("\nALERTA: Señal SHORT MUY FUERTE - Considerar abrir posición corta")
elif ultima_señal == 3:
    print("\nPRECAUCIÓN: Señal SHORT moderada - Esperar confirmación")
else:
    print("\nSin señal SHORT clara en este momento")
```
